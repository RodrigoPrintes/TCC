```python
import numpy as np

def velocidades_media(df_dir):

    media_frames = 0
    numero_frames_true = 0
    distance_lr = 18
    distance_up = 35

    dict_velocidades = {}

    media_velocidades = {
        'car': [0,0,0,10000,0],
        'motorcycle': [0,0,0,100000,0],
        'truck': [0,0,0,10000,0],
        ' bus': [0,0,0,10000,0],
        'veloc_total':[0]
    }

    for index, row in df_dir.iterrows():

        if  not pd.isna(row['Frame Out'] ):
            n_frames = row['Frame Out'] - row['Frame In']
            media_frames += n_frames
            numero_frames_true+=1

            dict_velocidades[row['ID']] = [row['Type'],n_frames]

        else:

            dict_velocidades[row['ID']] = [row['Type'],0]


    media_frames = media_frames/numero_frames_true


    media_total = 0
    total_v = 0
    new_dict_velocidades = {chave: valor for chave, valor in dict_velocidades.items() if valor[1] != 0}



    for key, value in new_dict_velocidades.items():

        tipo_veiculo = value[0]

        velocidade = distance_lr/(value[1]*time_frame)


        media_velocidades[tipo_veiculo][0] += velocidade
        media_total+=velocidade
        total_v +=1


        if velocidade < media_velocidades[tipo_veiculo][3]:
           media_velocidades[tipo_veiculo][3] = velocidade

        if velocidade > media_velocidades[tipo_veiculo][4]:
           media_velocidades[tipo_veiculo][4] = velocidade

        media_velocidades[tipo_veiculo][1] += 1
        media_velocidades[tipo_veiculo][2] = (media_velocidades[tipo_veiculo][0]/media_velocidades[tipo_veiculo][1])*3.6

    media_velocidades['veloc_total'][0] = media_total/total_v

    print("Velocidade Total: N_Veiculos; Media_VEL;VEL_MIN,VEL_MAX;MEDIA_TOTAL")
    print(media_velocidades)

    return media_velocidades

```
